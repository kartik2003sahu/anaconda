```python
tips=sns.load_dataset('tips')
```

```python
tips.describe()
```

```python
tips['day'].value_counts()
```

```python
tips.groupby(['sex','time'])['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='day',data=tips,hue='sex')
```

```python
sns.set_style('whitegrid')
```

```python
sns.barplot(x='day',y='total_bill',data=tips , estimator=sum)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='time', estimator=np.mean)
```

```python
sns.boxplot(x='size',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='smoker')
```
