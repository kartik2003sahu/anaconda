# Multivariate Variable

```python
for col in cat_cols:
    if col=='Product':
        continue
    for ncol in num_cols:
        pit.figure(figsize=(10,3))
        pit.suptitle(f'{col} vs {ncol}')
        pit.subplot(1,3,1)
        sns.boxplot(
            data=data,x=col,y=ncol,
            hue='Product',palette=choice(p_list))
        pit.subplot(1,3,2)
        sns.violinplot(
            data=data,x=col,y=ncol,
            hue='Product',palette=choice(p_list))
        pit.subplot(1,3,3)
        sns.barplot(
            data=data,x=col,y=ncol,
            hue='Product',palette=choice(p_list))
        pit.show()    
```
